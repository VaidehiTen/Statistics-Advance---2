1. **What is hypothesis testing in statistics?**
- Hypothesis testing is a statistical method used to determine if there is enough evidence in a sample data to draw conclusions about a population.
- It involves formulating two competing hypotheses, `the null hypothesis (H0)` and `the alternative hypothesis (Ha)`, and then collecting data to assess the evidence.
- Hypothesis testing helps determine if observed differences in sample data are likely due to a real effect in the population or just random chance.
- It's a decision-making process based on probabilities, allowing researchers to draw conclusions about populations based on sample evidence.

2. **What is the null hypothesis, and how does it differ from the alternative hypothesis?**
- The null hypothesis is the claim that there’s no effect in the population.
- If the sample provides enough evidence against the claim that there’s no effect in the population (p ≤ α), then we can reject the null hypothesis. Otherwise, we fail to reject the null hypothesis.
- Although “fail to reject” may sound awkward, it’s the only wording that statisticians accept. Be careful not to say you “prove” or “accept” the null hypothesis.
- Null hypotheses often include phrases such as “no effect,” “no difference,” or “no relationship.” When written in mathematical terms, they always include an equality (usually =, but sometimes ≥ or ≤).
-You can never know with complete certainty whether there is an effect in the population. Some percentage of the time, your inference about the population will be incorrect. When you incorrectly reject the null hypothesis, it’s called a type I error. When you incorrectly fail to reject it, it’s a type II error.

 **How does null hypothesis differ from the alternative hypothesis :**

- The null hypothesis assumes no effect or difference, while the alternative hypothesis assumes an effect or difference is present.
- Researchers typically aim to reject the null hypothesis and find support for the alternative hypothesis.
- The null hypothesis usually uses an equality symbol, while the alternative hypothesis uses inequality symbols.
- The null hypothesis provides a basis for comparison, while the alternative hypothesis is the statement being tested against it.
- Rejecting the null hypothesis suggests evidence supports the alternative hypothesis and implies a statistically significant result.

3. **What is the significance level in hypothesis testing, and why is it importan?**
- In hypothesis testing, `the significance level`, denoted by the Greek letter `α (alpha)`. A significance level is the probability set by the researcher to determine the threshold at which the null hypothesis is either accepted or rejected. It represents the likelihood of making a type I error and is typically set at 1%, 5%, or 10% based on various factors such as research type, sample size, and associated costs.

  **Why it's important :**

- *Controls Type I Errors*: The significance level determines the probability of making a Type I error (a false positive). Setting a low alpha value (e.g., 0.05 or 0.01) reduces the risk of falsely rejecting the null hypothesis.
- *Sets the Decision Rule*: The chosen significance level helps establish the critical region or critical value(s) in a hypothesis test.
  - If the p-value (calculated probability of observing the data given the null hypothesis is true) is less than or equal to the significance level, the null hypothesis is rejected, and the results are considered statistically significant.
  - If the p-value is greater than the significance level, the null hypothesis is not rejected.
- *Provides a Standard for Evaluation*: The significance level offers a benchmark for researchers to evaluate their findings and determine whether the observed effects are likely real or just due to chance.
- *Impacts Confidence Level*: Alpha is directly related to the confidence level of a study. For instance, an alpha of 0.05 corresponds to a 95% confidence level (1 - α).


4. **What does a P-value represent in hypothesis testing?**
- In hypothesis testing, a P-value represents the probability of obtaining results as extreme as, or more extreme than, the results observed in a study, assuming the null hypothesis is true.
- Essentially, it quantifies the likelihood that the observed data occurred by random chance, given that there's no real effect or difference being measured.
- A p-value is a statistical measurement used to validate a hypothesis against observed data.
- A p-value measures the probability of obtaining the observed results, assuming that the null hypothesis is true.
- The lower the p-value, the greater the statistical significance of the observed difference.

  

5. **How do you interpret the P-value in hypothesis testing?**
- The p-value, or probability value, is a fundamental concept in hypothesis testing used to assess the strength of evidence against a null hypothesis (H₀). The null hypothesis generally represents the absence of an effect or relationship you're investigating. The alternative hypothesis (H₁, or Hₐ) is what you're trying to prove – that a relationship exists or there's a difference between groups.

 **how to interpret the p-value:**

 > Definition: The p-value is the probability of obtaining results as extreme as, or more extreme than, the observed results, assuming the null hypothesis is true. In simpler terms, it's the likelihood that findings occurred by random chance if there were truly no effect or relationship in the population.

 > Comparison to the Significance Level (Alpha - α): Researchers set a significance level (alpha - α) before conducting a hypothesis test. The significance level is the maximum probability of rejecting the null hypothesis when it's actually true (Type I error). A common significance level is 0.05, meaning there's a 5% chance of making a Type I error.

 > Decision rule:
  - If the p-value ≤ α: The results are considered statistically significant, and the null hypothesis is rejected in favor of the alternative hypothesis. This indicates strong evidence against the null hypothesis, suggesting that the observed effect is unlikely to be due to chance.
  - If the p-value > α: The results are not considered statistically significant, and the null hypothesis is not rejected. There is insufficient evidence to conclude a significant effect, but it does not mean that the null hypothesis is true, according to Simply Psychology. There simply isn't enough evidence to rule out chance as an explanation.

 > Strength of evidence: A smaller p-value signifies stronger evidence against the null hypothesis. For instance, a p-value of 0.01 provides stronger evidence against the null hypothesis than a p-value of 0.04.

6. **What are Type 1 and Type 2 errors in hypothesis testing?**
- In statistical hypothesis testing, Type I and Type II errors represent incorrect conclusions drawn about a population based on sample data.

 **Type I error**

- A Type I error means rejecting the null hypothesis when it’s actually true. It means concluding that results are statistically significant when, in reality, they came about purely by chance or because of unrelated factors.
- The risk of committing this error is the significance level (alpha or α) you choose. That’s a value that you set at the beginning of your study to assess the statistical probability of obtaining your results (p value).
- The significance level is usually set at 0.05 or 5%. This means that your results only have a 5% chance of occurring, or less, if the null hypothesis is actually true.
- If the p value of your test is lower than the significance level, it means your results are statistically significant and consistent with the alternative hypothesis. If your p value is higher than the significance level, then your results are considered statistically non-significant.
- To reduce the Type I error probability, you can simply set a lower significance level.

 **Type II error**

- A Type II error means not rejecting the null hypothesis when it’s actually false. This is not quite the same as “accepting” the null hypothesis, because hypothesis testing can only tell you whether to reject the null hypothesis.
- Instead, a Type II error means failing to conclude there was an effect when there actually was. In reality, your study may not have had enough statistical power to detect an effect of a certain size.
- Power is the extent to which a test can correctly detect a real effect when there is one. A power level of 80% or higher is usually considered acceptable.
- The risk of a Type II error is inversely related to the statistical power of a study. The higher the statistical power, the lower the probability of making a Type II error.
- To (indirectly) reduce the risk of a Type II error, you can increase the sample size or the significance level.


7. **What is the difference between a one-tailed and a two-tailed test in hypothesis testing?**
- In hypothesis testing, both one-tailed and two-tailed tests are used to determine whether there is enough evidence to reject a null hypothesis. Here's a breakdown of the key differences:

 **One-tailed test**

- Directional Hypothesis: Used when the research hypothesis specifies a direction for the effect or relationship between variables.
- Alternative Hypothesis: States that the parameter being investigated is either greater than or smaller than a specific value (e.g., the new drug is more effective than the old drug).
- Critical Region: Located in one tail of the probability distribution (either left or right), representing the area where the null hypothesis would be rejected.
- Statistical Power: Potentially higher than a two-tailed test, as the power is concentrated in one specific direction.
- Use Cases: Appropriate when there is strong theoretical reason or prior evidence to predict a specific direction of effect, according to Towards Data Science. Examples include testing if a new drug improves recovery rates or if a new marketing campaign increases sales.  

 **Two-tailed test**

- Non-Directional Hypothesis: Used when the research hypothesis does not specify a direction for the effect or relationship.
- Alternative Hypothesis: States that the parameter being investigated is simply different from a specific value (e.g., the new treatment is different from the standard treatment).
- Critical Regions: Split between both tails of the probability distribution, allowing for the detection of differences in either direction (positive or negative).
- Statistical Power: Lower than a one-tailed test for the same significance level and sample size, as the power is divided between both tails.
- Use Cases: Appropriate when there is no specific expectation about the direction of the effect or when detecting any type of difference, whether positive or negative, is important. Examples include comparing two different medical treatments or evaluating competing marketing strategies where the outcome direction is unknown.

8. **What is the Z-test, and when is it used in hypothesis testing?**
- A Z-test compares one mean against a hypothesized value. It tests whether two means are the same. The data must approximately fit a normal distribution, or the test won't work. Parameters such as variance and standard deviation should be calculated when performing a Z-test.

 **When to use the Z-test**

- Large sample size:  Typically, a sample size greater than 30 is considered large enough to apply the Z-test.  This is based on the Central Limit Theorem, which states that the distribution of sample means approaches a normal distribution as the sample size increases, regardless of the population's distribution.
- Known population standard deviation: The Z-test requires knowing the population standard deviation (\(\sigma \)) or variance (\(\sigma ^{2}\)).
- Normally distributed data: The Z-test assumes that the population from which the samples are drawn is approximately normally distributed. This assumption is less critical with larger sample sizes due to the Central Limit Theorem, which allows the sampling distribution of the mean to be approximated as normal even if the underlying data isn't perfectly normal.
- Independent data points: All data points must be independent of one another.
- Equal variances for each sample: When comparing two samples, the Z-test assumes that the variances of the populations from which the samples are drawn are equal.

9. **How do you calculate the Z-score, and what does it represent in hypothesis testing?**
- Calculating the Z-score and its role in hypothesis testing
A Z-score, also known as a standard score, quantifies how many standard deviations a particular data point is away from the mean of its distribution. It's a crucial tool in statistics, particularly in hypothesis testing. According to Investopedia, a Z-score reveals if a value is typical or atypical for a dataset.

 **Z-score formula**

 The formula for calculating a Z-score is:

 z = (x - μ) / σ

 Where:

 z = Z-score

 x = The value being evaluated (the data point)

 μ = The population mean (average)

 σ = The population standard deviation (measure of data spread)

 **Here's what it represents:**

- Standardization: The Z-score standardizes the data. This makes it possible to compare results from different datasets with different means and standard deviations.
- Deviation from the mean: The Z-score indicates how far a sample mean or individual observation deviates from a hypothesized population mean. This is measured in standard deviations.
- Probability: By comparing the calculated Z-score to critical values from a standard normal distribution table or using statistical software, you can determine the probability of obtaining the observed results if the null hypothesis is true.
- Decision-making: Based on this probability (p-value) and a chosen significance level (alpha), you decide whether to reject or fail to reject the null hypothesis. If the p-value is less than or equal to the significance level, you reject the null hypothesis. This indicates a statistically significant difference.

10. **What is the T-distribution, and when should it be used instead of the normal distribution?**
- The t-distribution is a type of normal distribution that is used for smaller sample sizes. Normally-distributed data form a bell shape when plotted on a graph, with more observations near the mean and fewer observations in the tails.

 **The t-distribution should be used instead of the normal distribution in the following cases:**

- Small Sample Sizes: The t-distribution is most useful when dealing with small sample sizes, typically fewer than 30, where the Central Limit Theorem might not fully apply. The t-distribution accounts for increased variability when the sample size is smaller.
- Unknown Population Standard Deviation: When the population standard deviation (\(\sigma \)) is unknown, and it must be estimated using the sample standard deviation (s), the t-distribution is the appropriate choice. The t-distribution adjusts for the additional uncertainty introduced by estimating \(\sigma \) from the sample. If the population standard deviation is known, even with a small sample, the normal distribution can be used.
- Approximation to Normal Distribution: The t-distribution approaches the normal distribution as the sample size increases. If the sample size is greater than 30, the t-distribution is very close to the normal distribution, and either can be used. However, the t-distribution is more accurate if the population standard deviation is unknown.

11. **What is the difference between a Z-test and a T-test?**
- When analyzing data to compare means, two frequently used statistical tests are the z-test and the t-test. While both tests are used for hypothesis testing, their application depends primarily on whether the population standard deviation is known and the sample size.

 **Z-test**

- When to use: Use a Z-test when you know the population standard deviation (σ) OR when your sample size is large (typically n ≥ 30), even if you don't know the population standard deviation.
- Purpose: To determine if a sample mean differs significantly from a population mean, or if there's a significant difference between two independent sample means.
- Test Statistic: Uses the z-statistic, which follows a standard normal (Z) distribution.
- Key Assumptions:
  - Data is normally distributed.
  - Population standard deviation is known (or estimated reliably with a large sample).
  - Sample is random and observations are independent.

 **T-test**

- When to use: Use a T-test when the population standard deviation is unknown and your sample size is small (typically n < 30).
- Purpose: To determine if a sample mean differs significantly from a hypothesized population mean, or if there's a significant difference between two group means (either independent or paired samples).
- Test Statistic: Uses the t-statistic, which follows a t-distribution. The t-distribution has heavier tails than the normal distribution, accounting for the increased uncertainty when estimating the population standard deviation from a small sample.
- Degrees of freedom: For a one-sample t-test, the degrees of freedom are calculated as n-1.
- Key Assumptions:
  - Data in each group are approximately normally distributed.
  - Homogeneity of variances (for independent two-sample t-test, meaning the variances of the two groups are assumed to be equal).
  - Observations are independent.

12. **What is the T-test, and how is it used in hypothesis testing?**
- A `t test` is a statistical test that is used to compare the means of two groups. It is often used in hypothesis testing to determine whether a process or treatment actually has an effect on the population of interest, or whether two groups are different from one another.

 **How the t-test is used in hypothesis testing**

 Hypothesis testing is a statistical method for evaluating assumptions about a population using sample data. The t-test is integrated into this process as follows:

- Interpreting the Results: The t-test yields a p-value, indicating the likelihood of observing the measured difference if the null hypothesis is true. This is compared to a significance level (\(\alpha \)), typically 0.05, to make a decision. If the p-value is less than or equal to \(\alpha \), the null hypothesis is rejected, suggesting a significant difference. Otherwise, you fail to reject the null hypothesis, implying no significant difference.

  - Distribution: The Z-test statistic follows a standard normal distribution, whereas the T-test statistic follows a Student's T-distribution with degrees of freedom (\(n-1\)).
  - Robustness: The T-test is generally more robust to violations of the normality assumption, especially when sample sizes are small,

14. **What is a confidence interval, and how is it used to interpret statistical results?**
- In statistics, a confidence interval is a range of values likely to contain a population parameter that is unknown. Analysts often use confidence level values of 95% or 99% when calculating confidence intervals. Thus, if a point estimate is generated from a statistically significant population with a mean of 10.00 using a 95% confidence interval of 9.50 to 10.50, it means one is 95% confident that the true value from the population falls within that range.

 **Components of a confidence interval**

- Point Estimate: The calculated value from your sample.
- Margin of Error: Defines the range around the point estimate, influenced by sample variability and confidence level.
- Confidence Level: The percentage (e.g., 95%) representing the proportion of confidence intervals from repeated samples expected to contain the true population parameter.

 **How is it used to interpret statistical results?**

- Estimating the Range: They provide a range where the true population parameter is likely to be found, such as being 95% confident the true average height is between 160-170 cm.
- Quantifying Uncertainty: CIs show the uncertainty of an estimate based on sample data; a wider interval means more uncertainty.
- Assessing Precision: A narrow CI indicates a more precise estimate.
- Hypothesis Testing: CIs can be used with hypothesis testing. If the interval for a mean difference includes zero, it suggests no significant difference. If it excludes zero, there is a statistically significant difference.
- Comparing Groups: Non-overlapping CIs for different groups suggest a statistically significant difference between them.

15. **What is the margin of error, and how does it affect the confidence interval?**
- Margin of error is a statistical term that represents the range of uncertainty or variability around an estimate or measurement. It quantifies the degree of confidence we can have in the estimate's accuracy. When conducting surveys or collecting data, surveying the entire population is often not feasible or practical. Instead, a sample is taken, and the results are used to make inferences about the larger population.
- The margin of error helps to account for the potential discrepancy between the sample and the population, leading to more accurate findings from your data analytics.

 **How the margin of error affects the confidence interval**

- he confidence interval is a range of values derived from sample data within which the true population parameter is likely to fall, with a specified level of confidence. The margin of error determines the width of the confidence interval.   - A smaller margin of error results in a narrower confidence interval, suggesting a more precise estimate.
- For example, if a poll estimates that 52% of people prefer a product with a ±3% margin of error at a 95% confidence level, the confidence interval would be between 49% and 55%. This means the user can be 95% confident that the true population percentage of those who prefer the product lies within this range.



16. **How is Bayes' Theorem used in statistics, and what is its significance?**
- Bayes' theorem is a fundamental concept in statistics that enables us to calculate and update the probability of an event happening based on new information or evidence. It is often referred to as the formula for "the probability of causes" because it helps us understand the likelihood of a particular cause given an observed event.

 **How is Bayes' theorem used?**

- At its core, Bayes' theorem allows for the revision of prior probabilities (initial beliefs) in light of new data to arrive at a more informed posterior probability (updated belief). The formula is expressed as:
  
  P(A|B)=frac{P(B|A)P(A)}{P(B)}

- The elements of the formula are defined as:
  
   P(A|B): posterior probability
  
   P(B|A): likelihood
   
   P(A): prior probability
   
   P(B): marginal probability
- The theorem is useful when prior knowledge needs to be refined with new data.

 **Significance of Bayes' theorem in statistics**

- Update beliefs and improve accuracy with new evidence.
- Quantify uncertainty, helpful with incomplete data.
- Integrate prior knowledge for comprehensive insights.
- Facilitate better decision-making under uncertainty.
- Compare hypotheses using Bayes factors.

18. **What is the Chi-square goodness of fit test, and how is it applied?**

 **The Chi-Square goodness-of-fit test**
- The Chi-Square goodness-of-fit test is a statistical hypothesis test that assesses how well observed categorical data fits an expected or hypothesized distribution. It is often used to determine if a sample of data is representative of a larger population or to test if a variable follows a particular theoretical distribution.

 **How it's applied**

- State the hypotheses: This includes the null hypothesis (H0) that the observed data fits the expected distribution and the alternative hypothesis (Ha) that it does not.
- Calculate the expected frequencies: These are the frequencies expected if the null hypothesis is true.
- Calculate the Chi-Square test statistic (χ²): This is done using the formula χ² = Σ [(Oi - Ei)² / Ei], comparing observed (Oi) and expected (Ei) frequencies.
- Determine the degrees of freedom (df): For this test, df is the number of categories (k) minus 1 (df = k - 1).
- Determine the critical value: This value is found using a Chi-Square distribution table or software, based on the degrees of freedom and significance level (α).
- Compare the calculated χ² to the critical value (or p-value to α): If χ² is greater than the critical value (or p-value < α), the null hypothesis is rejected. Otherwise, it is not rejected.   



19. **What is the F-distribution, and when is it used in hypothesis testing?**
- The F-distribution, also known as the Fisher-Snedecor distribution, is a continuous probability distribution frequently employed in hypothesis testing and the analysis of variance (ANOVA). It is named after Sir Ronald Fisher, a prominent statistician who played a key role in its development and application.

 **When is the F-distribution used in hypothesis testing**

- Comparing two population variances (F-test): The F-test evaluates if the variances of two independent samples from normal populations are significantly different. The null hypothesis posits that variances are equal (\(\sigma _{1}^{2}=\sigma _{2}^{2}\)), with the F-statistic being the ratio of the sample variances (\(F=s_{1}^{2}/s_{2}^{2}\)). Rejecting the null hypothesis occurs if the calculated F-value exceeds the critical value.
- Analyzing variance (ANOVA): ANOVA uses the F-distribution to test if the means of three or more groups differ significantly. The null hypothesis states all group means are equal. The F-statistic is the ratio of between-group variance to within-group variance. A high F-value suggests significant differences between group means relative to variability within groups.

20. **What is an ANOVA test, and what are its assumptions?**
- ANOVA, or Analysis of Variance, is a statistical technique used to compare the means of three or more groups simultaneously. It examines the variation within and between groups to determine if observed differences in group means are statistically significant, meaning they are unlikely to be due to random chance alone.
- The core principle behind ANOVA is to divide the total variance in a dataset into different components from various sources. Specifically, ANOVA considers:
  - Within-group variability: This reflects the inherent variability within each group, assuming the independent variable has no effect.
  - Between-group variability: This represents the differences between the means of the various groups, potentially caused by the independent variable or factors being studied.

 **Assumptions of ANOVA**   

- Independence of Observations: Each data point must be independent of other data points, both within and between groups. This usually means using random sampling or avoiding clustered data unless employing statistical methods that account for dependencies, such as repeated measures ANOVA.
- Normality: The distribution of the dependent variable within each group should be approximately normal. While ANOVA is somewhat robust to minor violations, particularly with larger sample sizes, extreme non-normality can affect the accuracy of p-values, according to Number Analytics.
- Homogeneity of Variance (Homoscedasticity): The variance of the dependent variable should be roughly equal across all groups being compared. Unequal variances can lead to an increased risk of Type I errors, especially with unequal sample sizes, says Number Analytics. The assumption can be checked using tests like Levene's test.  

21. **What are the different types of ANOVA tests?**
- ANOVA, or Analysis of Variance, is a statistical method used to compare the means of two or more groups to determine if there are significant differences between them. Several types of ANOVA tests are available, each designed for specific scenarios based on the number of independent and dependent variables, and whether repeated measures are involved.

 **Here are the main types of ANOVA tests:**

- **One-Way ANOVA:**
  - This is the most basic form of ANOVA, used when you have one categorical independent variable with three or more groups (or "levels") and a continuous dependent variable.
  - For example, you might use a one-way ANOVA to determine if there's a difference in average test scores among students taught using three different teaching methods (independent variable: teaching method; levels: method A, method B, method C; dependent variable: test score).
- **Two-Way ANOVA:**
  - This test is used when you have two independent variables (each with two or more levels) and a continuous dependent variable. Two-way ANOVA examines not only the main effects of each independent variable but also the interaction effect between them.
  - For example, you could use a two-way ANOVA to investigate the effect of both teaching method and study technique on student exam scores, and whether the effect of the teaching method depends on the study technique used.
- **Factorial ANOVA**:
  - This is an extension of the two-way ANOVA, used when there are two or more independent variables, each with multiple levels. Factorial ANOVA allows for the analysis of more complex designs and interactions between the independent variables.
  - For instance, a business might use a factorial ANOVA to examine the combined effects of age, income, and education level on consumer purchasing habits.
- **Repeated Measures ANOVA:**
  - This test is employed when the same group of participants is measured multiple times under different conditions or over a period of time. It's essentially an extension of the paired t-test for more than two repeated measurements.
  - For example, a researcher might measure a group of patients' blood pressure at baseline, 30 minutes, and 60 minutes after taking a medication to assess the change in blood pressure over time.  

22. **What is the F-test, and how does it relate to hypothesis testing?**
- **What is an F-test?**
  - The F-test essentially works by comparing the variances of two or more groups.
  - The underlying principle is that if the variances are equal, the ratio of the variances will be close to 1.
  - F-tests are often used in research studies to determine whether the difference in the means of two populations is statistically significant.

 **How it relates to hypothesis testing**  

- Hypothesis Formulation: The first step in hypothesis testing using an F-test is to state a null hypothesis (H₀) and an alternative hypothesis (H₁).
  - Null Hypothesis (H₀): This hypothesis states that there is no statistically significant difference between the variances of the groups being compared. In some cases, it states that there is no predictive relationship between variables in the population.
  - Alternative Hypothesis (H₁): This hypothesis proposes that there is a statistically significant difference between the variances or a predictive relationship exists.
- Calculating the F-statistic: The F-statistic is calculated as a ratio of two variances. For comparing two samples, the formula is: F = s₁²/s₂², where s₁² and s₂² are the variances of the two samples. To simplify calculations, the larger variance is typically placed in the numerator.
- Determining the Critical Value: The critical value of F is obtained from an F-distribution table, based on the degrees of freedom (number of observations minus 1 for each sample) and the chosen significance level (alpha, α).
- Decision Making: The calculated F-statistic is compared to the critical F-value:
  - If F-statistic > Critical F-value: Reject the null hypothesis. This indicates that the observed difference in variances or the relationship between variables is statistically significant.
  - If F-statistic < Critical F-value: Fail to reject the null hypothesis. This means that the observed difference is likely due to random chance and not statistically significant.

2. **Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).**

15. **Implement hypothesis testing using Z-statistics for a sample dataset.**

```python
import numpy as np
from scipy.stats import norm

def z_test(sample, mu_0, sigma, alpha=0.05, alternative='two-sided'):

    n = len(sample)
    sample_mean = np.mean(sample)

    # Calculate standard error
    std_error = sigma / np.sqrt(n)

    # Calculate Z statistic
    z_stat = (sample_mean - mu_0) / std_error

    # Calculate p-value based on alternative hypothesis
    if alternative == 'two-sided':
        p_value = 2 * (1 - norm.cdf(abs(z_stat)))
    elif alternative == 'less':
        p_value = norm.cdf(z_stat)
    elif alternative == 'greater':
        p_value = 1 - norm.cdf(z_stat)
    else:
        raise ValueError("alternative must be 'two-sided', 'less', or 'greater'")

    # Conclusion
    reject = p_value < alpha
    conclusion = 'Reject H0' if reject else 'Fail to reject H0'

    return z_stat, p_value, conclusion

# Example usage:
sample_data = [105, 102, 98, 110, 115, 108, 112, 99, 104, 101]
hypothesized_mean = 100
population_std_dev = 10  # known population std deviation
significance_level = 0.05

z_stat, p_val, conclusion = z_test(sample_data, hypothesized_mean, population_std_dev, significance_level)

print(f'Z-statistic: {z_stat:.3f}')
print(f'p-value: {p_val:.4f}')
print(f'Conclusion: {conclusion}')
```

16. **Create a confidence interval for a dataset using Python and interpret the result.**

```python
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for the mean of the data.

    Parameters:
    - data: list or array of sample data
    - confidence: confidence level (default 0.95 for 95%)

    Returns:
    - (lower_bound, upper_bound): tuple with confidence interval limits
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # standard error of the mean

    # Degrees of freedom
    df = n - 1

    # Critical t-value
    t_crit = stats.t.ppf((1 + confidence) / 2, df)

    # Margin of error
    margin = t_crit * std_err

    lower_bound = mean - margin
    upper_bound = mean + margin

    return lower_bound, upper_bound

# Example usage:
sample_data = [12, 15, 14, 16, 13, 14, 15, 14, 16, 17]
conf_level = 0.95

ci_lower, ci_upper = confidence_interval(sample_data, confidence=conf_level)
mean_val = np.mean(sample_data)

print(f"Sample mean: {mean_val:.2f}")
print(f"{int(conf_level*100)}% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
print(f"Interpretation: We are {int(conf_level*100)}% confident that the population mean lies within this interval.")
```

1. **Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.**

```python
import math
from scipy.stats import norm

def z_test(sample_mean, population_mean, population_std, sample_size, alpha=0.05):
    """
    Perform a one-sample Z-test.

    Parameters:
    - sample_mean: Mean of the sample
    - population_mean: Known population mean
    - population_std: Known population standard deviation
    - sample_size: Size of the sample
    - alpha: Significance level (default 0.05)

    Returns:
    - z_stat: Calculated Z statistic
    - p_value: Two-tailed p-value
    - interpretation: Text interpretation of the result
    """
    # Calculate standard error
    standard_error = population_std / math.sqrt(sample_size)

    # Calculate Z statistic
    z_stat = (sample_mean - population_mean) / standard_error

    # Calculate two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    # Interpretation
    if p_value < alpha:
        interpretation = ("Reject the null hypothesis: "
                          "There is a statistically significant difference "
                          "between the sample mean and the population mean.")
    else:
        interpretation = ("Fail to reject the null hypothesis: "
                          "There is no statistically significant difference "
                          "between the sample mean and the population mean.")

    return z_stat, p_value, interpretation


# Example usage:
sample_mean = 105
population_mean = 100
population_std = 15
sample_size = 30
alpha = 0.05

z_stat, p_value, interpretation = z_test(sample_mean, population_mean, population_std, sample_size, alpha)

print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Interpretation: {interpretation}")
```

```python
import math
from scipy.stats import norm

def one_sample_z_test(sample_mean, population_mean, population_std, sample_size, alpha=0.05, tail='two'):
    """
    Perform a one-sample Z-test.

    Parameters:
    - sample_mean: Mean of the sample
    - population_mean: Known population mean
    - population_std: Known population standard deviation
    - sample_size: Sample size
    - alpha: Significance level (default 0.05)
    - tail: 'two' for two-tailed test (default), 'left' or 'right' for one-tailed tests

    Returns:
    - z_stat: Calculated Z statistic
    - p_value: P-value of the test
    - conclusion: Interpretation of the test result
    """
    # Calculate standard error
    standard_error = population_std / math.sqrt(sample_size)

    # Calculate Z statistic
    z_stat = (sample_mean - population_mean) / standard_error

    # Calculate p-value based on tail type
    if tail == 'two':
        p_value = 2 * (1 - norm.cdf(abs(z_stat)))
    elif tail == 'right':
        p_value = 1 - norm.cdf(z_stat)
    elif tail == 'left':
        p_value = norm.cdf(z_stat)
    else:
        raise ValueError("tail must be 'two', 'left', or 'right'")

    # Interpretation
    if p_value < alpha:
        conclusion = "Reject the null hypothesis: significant difference detected."
    else:
        conclusion = "Fail to reject the null hypothesis: no significant difference detected."

    return z_stat, p_value, conclusion


# Example usage:
sample_mean = 102
population_mean = 100
population_std = 15
sample_size = 50
alpha = 0.05
tail = 'two'  # Can be 'two', 'left', or 'right'

z_stat, p_value, conclusion = one_sample_z_test(sample_mean, population_mean, population_std, sample_size, alpha, tail)

print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Conclusion: {conclusion}")
```

4. **Perform a two-tailed Z-test using Python and visualize the decision region on a plot.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import math

def two_tailed_z_test(sample_mean, population_mean, population_std, sample_size, alpha=0.05):
    # Calculate standard error
    se = population_std / math.sqrt(sample_size)

    # Calculate Z statistic
    z_stat = (sample_mean - population_mean) / se

    # Calculate p-value (two-tailed)
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    # Critical z-values for rejection regions
    z_critical = norm.ppf(1 - alpha/2)

    return z_stat, p_value, z_critical

def plot_z_test(z_stat, z_critical, alpha=0.05):
    # Setup x values for the standard normal distribution
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    plt.figure(figsize=(10,6))
    plt.plot(x, y, label='Standard Normal Distribution')

    # Shade rejection regions
    plt.fill_between(x, 0, y, where=(x <= -z_critical), color='red', alpha=0.5, label=f'Rejection Region\nZ < -{z_critical:.2f}')
    plt.fill_between(x, 0, y, where=(x >= z_critical), color='red', alpha=0.5, label=f'Rejection Region\nZ > {z_critical:.2f}')

    # Plot the test statistic
    plt.axvline(z_stat, color='blue', linestyle='--', linewidth=2, label=f'Z-statistic = {z_stat:.2f}')

    # Labels and title
    plt.title('Two-tailed Z-test Decision Regions')
    plt.xlabel('Z value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()


# Example parameters
population_mean = 100
population_std = 15
sample_size = 30
sample_mean = 110  # Example sample mean
alpha = 0.05

# Perform the test
z_stat, p_value, z_critical = two_tailed_z_test(sample_mean, population_mean, population_std, sample_size, alpha)

print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Critical Z value (two-tailed, alpha={alpha}): ±{z_critical:.4f}")

# Visualize
plot_z_test(z_stat, z_critical, alpha)
```

5. **Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import math

def plot_type_errors(pop_mean, pop_std, sample_size, alpha, alt_mean):
    """
    Plot Type I and Type II errors for a one-sample Z-test.

    Parameters:
    - pop_mean: Population mean under the null hypothesis (H0)
    - pop_std: Population standard deviation (assumed known)
    - sample_size: Sample size
    - alpha: Significance level (Type I error rate)
    - alt_mean: True mean under alternative hypothesis (H1)
    """

    # Calculate standard error
    se = pop_std / math.sqrt(sample_size)

    # Critical Z value for given alpha (two-tailed test)
    z_critical_low = norm.ppf(alpha / 2)
    z_critical_high = norm.ppf(1 - alpha / 2)

    # Critical sample means defining rejection regions under H0
    crit_low = pop_mean + z_critical_low * se
    crit_high = pop_mean + z_critical_high * se

    # Calculate beta (Type II error): Probability of failing to reject H0 when H1 is true
    beta_low = norm.cdf(crit_low, loc=alt_mean, scale=se)
    beta_high = 1 - norm.cdf(crit_high, loc=alt_mean, scale=se)
    beta = beta_low + beta_high

    # Plot range covering both distributions well
    x_min = min(pop_mean - 4*se, alt_mean - 4*se)
    x_max = max(pop_mean + 4*se, alt_mean + 4*se)
    x = np.linspace(x_min, x_max, 1000)

    # Distributions under H0 and H1
    h0_pdf = norm.pdf(x, loc=pop_mean, scale=se)
    h1_pdf = norm.pdf(x, loc=alt_mean, scale=se)

    plt.figure(figsize=(12,6))

    # Plot PDFs
    plt.plot(x, h0_pdf, label='Null Hypothesis (H0)', color='blue')
    plt.plot(x, h1_pdf, label='Alternative Hypothesis (H1)', color='green')

    # Shade rejection regions under H0 (Type I error areas)
    plt.fill_between(x, 0, h0_pdf, where=(x <= crit_low), color='red', alpha=0.3, label='Type I Error Region (α)')
    plt.fill_between(x, 0, h0_pdf, where=(x >= crit_high), color='red', alpha=0.3)

    # Shade acceptance region under H1 (Type II error area)
    plt.fill_between(x, 0, h1_pdf, where=(x > crit_low) & (x < crit_high), color='orange', alpha=0.3, label='Type II Error Region (β)')

    # Plot critical lines
    plt.axvline(crit_low, color='red', linestyle='--', label=f'Critical Value Lower ({crit_low:.2f})')
    plt.axvline(crit_high, color='red', linestyle='--', label=f'Critical Value Upper ({crit_high:.2f})')

    plt.title('Type I (α) and Type II (β) Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Significance level (Type I error, α): {alpha:.4f}")
    print(f"Type II error (β) for true mean {alt_mean}: {beta:.4f}")
    print(f"Power of the test (1 - β): {1 - beta:.4f}")

# Example usage:
plot_type_errors(
    pop_mean=100,       # Mean under null hypothesis
    pop_std=15,         # Known population std dev
    sample_size=40,     # Sample size
    alpha=0.05,         # Significance level
    alt_mean=110        # True mean under alternative hypothesis
)
```

6. **Write a Python program to perform an independent T-test and interpret the results.**


```python
import numpy as np
from scipy.stats import ttest_ind

def independent_t_test(sample1, sample2, alpha=0.05):
    """
    Perform an independent two-sample t-test and interpret the results.

    Parameters:
    - sample1: list or array of sample values for group 1
    - sample2: list or array of sample values for group 2
    - alpha: significance level (default 0.05)

    Returns:
    - t_stat: t statistic
    - p_value: p-value of the test
    - interpretation: text conclusion of the test
    """
    t_stat, p_value = ttest_ind(sample1, sample2, equal_var=False)  # Welch's t-test by default

    if p_value < alpha:
        interpretation = ("Reject the null hypothesis: "
                          "There is a statistically significant difference "
                          "between the two group means.")
    else:
        interpretation = ("Fail to reject the null hypothesis: "
                          "There is no statistically significant difference "
                          "between the two group means.")

    return t_stat, p_value, interpretation

# Example usage
np.random.seed(0)

# Generate two random samples with different means
group1 = np.random.normal(loc=50, scale=10, size=30)
group2 = np.random.normal(loc=55, scale=10, size=30)

t_stat, p_value, interpretation = independent_t_test(group1, group2)

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Interpretation: {interpretation}")
```

7. **Perform a paired sample T-test using Python and visualize the comparison results.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel

def paired_t_test(group1, group2, alpha=0.05):
    """
    Perform a paired t-test and interpret the results.

    Parameters:
    - group1: array-like, measurements before
    - group2: array-like, measurements after
    - alpha: significance level (default 0.05)

    Returns:
    - t_stat: t statistic
    - p_value: p-value
    - interpretation: text interpretation
    """
    t_stat, p_value = ttest_rel(group1, group2)

    if p_value < alpha:
        interpretation = "Reject null hypothesis: Significant difference between paired samples."
    else:
        interpretation = "Fail to reject null hypothesis: No significant difference detected."

    return t_stat, p_value, interpretation

def plot_paired_data(group1, group2):
    """
    Visualize paired data with lines connecting paired observations.
    """
    n = len(group1)
    x = [1, 2]

    plt.figure(figsize=(8,6))

    # Plot paired points and connect with lines
    for i in range(n):
        plt.plot(x, [group1[i], group2[i]], marker='o', color='gray', alpha=0.6)

    # Plot means with bigger markers
    mean1, mean2 = np.mean(group1), np.mean(group2)
    plt.scatter([1], [mean1], color='blue', s=100, label=f'Mean Before: {mean1:.2f}')
    plt.scatter([2], [mean2], color='red', s=100, label=f'Mean After: {mean2:.2f}')

    plt.xticks([1, 2], ['Before', 'After'])
    plt.ylabel('Measurement')
    plt.title('Paired Sample Data Visualization')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage:
np.random.seed(1)
before = np.random.normal(loc=50, scale=5, size=25)
after = before + np.random.normal(loc=2, scale=5, size=25)  # Simulate some change

t_stat, p_value, interpretation = paired_t_test(before, after)
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Interpretation: {interpretation}")

plot_paired_data(before, after)
```

9. **Write a Python function to calculate the confidence interval for a sample mean and explain its significance.**

```python
import numpy as np
from scipy.stats import t

def confidence_interval(sample, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
    - sample: list or numpy array of sample data
    - confidence: confidence level (default 0.95 for 95% CI)

    Returns:
    - (lower_bound, upper_bound): tuple representing the confidence interval
    """
    n = len(sample)
    mean = np.mean(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)  # Standard error of the mean
    df = n - 1  # degrees of freedom

    # t critical value for two-tailed test
    t_crit = t.ppf((1 + confidence) / 2, df)

    margin_of_error = t_crit * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return lower_bound, upper_bound

# Example usage:
import numpy as np
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=40)

ci_lower, ci_upper = confidence_interval(sample_data, confidence=0.95)
print(f"95% confidence interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

10. **Write a Python program to calculate the margin of error for a given confidence level using sample data.**

```python
import numpy as np
from scipy.stats import t

def margin_of_error(sample, confidence=0.95):
    """
    Calculate the margin of error for the sample mean at the specified confidence level.

    Parameters:
    - sample: list or numpy array of sample data
    - confidence: confidence level (default 0.95 for 95%)

    Returns:
    - margin_of_error: the margin of error
    """
    n = len(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)  # standard error of the mean
    df = n - 1  # degrees of freedom
    t_crit = t.ppf((1 + confidence) / 2, df)  # two-tailed t critical value

    return t_crit * std_err

# Example usage:
import numpy as np
np.random.seed(123)
sample_data = np.random.normal(loc=50, scale=8, size=35)

moe = margin_of_error(sample_data, confidence=0.95)
print(f"Margin of Error at 95% confidence: {moe:.3f}")
```

11. **Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.**

```python
def bayes_theorem(prior, likelihood, false_positive_rate):
    """
    Calculate the posterior probability P(H|E) using Bayes' theorem.

    Parameters:
    - prior: P(H), prior probability of hypothesis (e.g., disease prevalence)
    - likelihood: P(E|H), probability of evidence given hypothesis (true positive rate)
    - false_positive_rate: P(E|¬H), probability of evidence given hypothesis is false (false positive rate)

    Returns:
    - posterior: P(H|E), posterior probability of hypothesis given evidence
    """
    # Total probability of evidence P(E)
    evidence = likelihood * prior + false_positive_rate * (1 - prior)

    # Bayes' theorem formula
    posterior = (likelihood * prior) / evidence
    return posterior

# Example usage:
prior_disease = 0.01              # 1% prevalence
true_positive_rate = 0.99         # 99% sensitivity
false_positive_rate = 0.05        # 5% false positive rate

posterior_prob = bayes_theorem(prior_disease, true_positive_rate, false_positive_rate)
print(f"Posterior probability of having disease given positive test: {posterior_prob:.4f}")
```

 **Explanation:**

- We start with a prior belief about disease prevalence (e.g., 1%).
- We observe a positive test result (the evidence).
- We update our belief (calculate posterior) using Bayes’ theorem, factoring in test accuracy and false positive rates.
- Despite a positive test, the posterior probability might still be relatively low if the disease is rare.

12. **Perform a Chi-square test for independence between two categorical variables in Python.**

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Example contingency table:
# Suppose we have data on two categorical variables: Gender and Preference
# Rows: Gender (Male, Female)
# Columns: Preference (Product A, Product B)

data = np.array([[30, 10],   # Male: 30 prefer A, 10 prefer B
                 [20, 40]])  # Female: 20 prefer A, 40 prefer B

# Or equivalently as a pandas DataFrame for better readability
df = pd.DataFrame(data, index=['Male', 'Female'], columns=['Product A', 'Product B'])

print("Contingency Table:")
print(df)

# Perform Chi-square test for independence
chi2, p, dof, expected = chi2_contingency(data)

print(f"\nChi-square statistic: {chi2:.4f}")
print(f"Degrees of freedom: {dof}")
print(f"P-value: {p:.4f}")

print("\nExpected frequencies if independent:")
print(pd.DataFrame(expected, index=df.index, columns=df.columns))

# Interpretation
alpha = 0.05
if p < alpha:
    print("\nResult: Reject the null hypothesis — variables are dependent (association exists).")
else:
    print("\nResult: Fail to reject the null hypothesis — no evidence of association (variables independent).")
```

13. **Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.**

```python
import numpy as np

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for Chi-square test.

    Parameters:
    - observed: 2D numpy array or list of lists with observed frequencies

    Returns:
    - expected: 2D numpy array with expected frequencies
    """
    observed = np.array(observed)
    row_sums = observed.sum(axis=1).reshape(-1, 1)  # sum of each row (column vector)
    col_sums = observed.sum(axis=0).reshape(1, -1)  # sum of each column (row vector)
    total = observed.sum()

    expected = (row_sums @ col_sums) / total
    return expected

# Example observed data
observed_data = [
    [30, 10],
    [20, 40]
]

expected_freq = calculate_expected_frequencies(observed_data)
print("Observed Frequencies:\n", np.array(observed_data))
print("\nExpected Frequencies:\n", expected_freq)
```

14. **Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.**

```python
import numpy as np
from scipy.stats import chisquare

# Observed frequencies (example: counts of categories)
observed = np.array([50, 30, 20])

# Expected frequencies under the null hypothesis (must sum to same total as observed)
# For example, expecting equal proportions among 3 categories:
expected = np.array([33.33, 33.33, 33.34])

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print(f"Chi-square statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: Observed data does NOT fit the expected distribution.")
else:
    print("Fail to reject null hypothesis: Observed data fits the expected distribution.")
```
